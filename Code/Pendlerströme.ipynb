```python
sg = next(g.subgraph(c) for c in nx.connected_components(g))
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(sg, ax=ax,
                 node_size=5,
                 font_size=6, alpha=.5,
                 width=.5)
ax.set_axis_off()
```
